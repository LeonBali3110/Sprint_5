# Ejercicio 1
Descarga el dataset adjunto de datos oficiales de la UEFA y selecciona un atributo del conjunto de datos. Calcula el p-valor y di si rechaza la hipótesis nula cogiendo un alfa de 5%. 

H0 = la distribución de datos es Gaussiana

H1 = la distribución de datos es no Gaussiana

```python
from scipy.stats import shapiro
stat, p = shapiro(df['MP'])
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
 print('Probablemente Gaussiana')
else:
 print('Probablemente no Gaussiana')
```

```python
import seaborn as sns
sns.histplot(df['MP'], kde = True)
```

Conclusión: Rechazamos la hipotesis nula. La muestra no tiene una distrubución normal. Lo que quiere decir que hay equipos que jugaron muchos partidos mientras que otros jugaron muy pocos.

# Ejercicio 2
Con el mismo dataset selecciona otros dos atributos del conjunto de datos. Calcula los p-valores y di si rechazan la hipótesis nula cogiendo un alfa de 5%.

H0: las dos muestras son independentes.

H1: hay una dependencia entre las dos muestras.

```python
stat, p = shapiro(df['Pts'])
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
 print('Probablemente Gaussiana')
else:
 print('Probablemente no Gaussiana')
```

```python
stat, p = shapiro(df['W'])
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
 print('Probablemente Gaussiana')
else:
 print('Probablemente no Gaussiana')
```

```python
sns.histplot(df['W'], kde = True)
```

```python
sns.histplot(df['Pts'], kde = True)
```

Para poder utilizar el metodo spearmanr es necesario que ambas sean del mismo tipo de distribución.

```python
from scipy.stats import spearmanr
stat, p = spearmanr(df['Pts'], df['W'])
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
	print('Probablemente independente')
else:
	print('Probablemente dependente')
```

Conclusión: Rechazamos la hipotesis nula. Es probable que las Vicorias esten relacionadas con los Puntos que tiene cada equipo. Tiene sentido ya que ganar un partido te da puntos.

# Ejercicio 3
Continúa con el conjunto de datos adjunto y selecciona tres atributos del conjunto de datos. Calcula el p-valor y di si rechaza la hipótesis nula cogiendo un alfa de 5%.

H0: el promedio de las muestras son iguales.

H1: uno o más de los promedios de las muestras no son iguales.

```python
from scipy.stats import f_oneway
stat, p = f_oneway(df['L'], df['GA'], df['xGA'])
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
 print('Probablemente la misma distribución')
else:
 print('Probablemente diferentes distribuciones')
```

```python
sns.histplot(df['GA'], kde = True)
sns.histplot(df['L'], kde = True)
sns.histplot(df['xGA'], kde = True)
```

Conclusión: Rechazamos la hipotesis nula. Los promedios de una de las muestras no es igual. Tal vez si excluyeramos la variable erronea el resultado seria diferente. Pero entre las estas tres no tienen promedios igualmente distribuidos.
